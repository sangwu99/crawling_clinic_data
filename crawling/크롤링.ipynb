```python
import selenium
from selenium import webdriver
from selenium.webdriver import ActionChains

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import Select
from time import sleep
```

```python
URL = 'http://opendata.hira.or.kr/op/opc/olapDiagBhvInfo.do'
driver = webdriver.Chrome(executable_path='chromedriver')
driver.get(url=URL)
```

```python
data = pd.read_excel('행위코드_소재지.xls',sheet_name='행위코드')
Code = data['행위']
Code = Code[1:194]
Code.drop_duplicates(inplace=True)
Code = Code.reset_index(drop=True)
```

```python
wrong_code = []
df = pd.DataFrame([])
path = "엑셀/"

for i in range(0,len(Code)):
    try:
        window = driver.find_element(By.ID,'searchPopup')
        window.click()
        driver.implicitly_wait(10)

        driver.switch_to.window(driver.window_handles[-1])
        driver.implicitly_wait(10)

        searchbox = driver.find_element(By.ID,'searchWrd1')
        driver.implicitly_wait(10)

        searchbox.send_keys(Code[i])
        driver.implicitly_wait(10)

        searchbox.send_keys(Keys.ENTER)
        driver.implicitly_wait(10)

        
        care = driver.find_element(By.CLASS_NAME,'select')
        driver.implicitly_wait(10)

        care.click()
        driver.implicitly_wait(10)

        driver.switch_to.window(driver.window_handles[0])
        driver.implicitly_wait(10)

        
        organization = driver.find_element(By.LINK_TEXT,'요양기관소재지별')

        organization.click()
        driver.implicitly_wait(10)
        sleep(5)

        
        content = driver.find_element(By.TAG_NAME,'iframe')
        driver.implicitly_wait(10)

        
        driver.switch_to.frame(content)
        driver.implicitly_wait(10)

        sleep(3)
        try:
            years = driver.find_element(By.XPATH,'//*[@id="ext-gen1645"]/table/tbody/tr/td/div[6]/select')
            driver.implicitly_wait(10)

            
            selectyear = Select(years)
            
            try:
                selectyear.select_by_value('2010')
            except:
                try:
                    selectyear.select_by_value('2014')
                except:
                    pass
                
            driver.implicitly_wait(10)
            sleep(2)
        except:
            pass

        view = driver.find_element(By.XPATH,'//*[@id="ext-gen1645"]/table/tbody/tr/td/div[9]/span')
        
        sleep(2)

        view.click()
        driver.implicitly_wait(10)

        sleep(30)
        
        downloadexcel = driver.find_element(By.XPATH,'//*[@id="panel-1104-body"]/div[3]/div[1]/div[1]/div[2]/div[1]')
        downloadexcel.click()
        
        sleep(5)
        
        list_file = glob.glob(path + "*.xls")
        
        try:
            
            if len(list_file) == 1:
                data = pd.read_excel(list_file[0],header=5)        
                
            else:
                data = pd.read_excel('엑셀\\2_진료행위요양기관소재지별현황(심사년도) ({i}).xls'.format(i=i),header=5)
            
            try:
                for num in range(1,13):
                    baseline = data.iloc[:,0:2]
                    
                    plusline = data.iloc[:,(num*3-1):(num*3+2)]
                    
                    middle = pd.concat([baseline,plusline],axis=1)
                    middle['년도'] = middle.iloc[0,3]
                    middle.iloc[1,-1] = '년도'
                    header = middle.iloc[1]
                    middle.rename(columns=header,inplace=True)
                    middle = middle.iloc[2:]

                    df = pd.concat([df,middle])
            except:
                try:
                    for num in range(1,9):
                        baseline = data.iloc[:,0:2]
                        
                        plusline = data.iloc[:,(num*3-1):(num*3+2)]
                        
                        middle = pd.concat([baseline,plusline],axis=1)
                        middle['년도'] = middle.iloc[0,3]
                        middle.iloc[1,-1] = '년도'
                        header = middle.iloc[1]
                        middle.rename(columns=header,inplace=True)
                        middle = middle.iloc[2:]

                        df = pd.concat([df,middle])
                except:
                    try:
                        for num in range(1,6):
                            baseline = data.iloc[:,0:2]
                            
                            plusline = data.iloc[:,(num*3-1):(num*3+2)]
                            
                            middle = pd.concat([baseline,plusline],axis=1)
                            middle['년도'] = middle.iloc[0,3]
                            middle.iloc[1,-1] = '년도'
                            header = middle.iloc[1]
                            middle.rename(columns=header,inplace=True)
                            middle = middle.iloc[2:]

                            df = pd.concat([df,middle])
                    except:
                        pass

        except:
            pass              
            
        driver.switch_to.default_content()
        
        driver.implicitly_wait(10)
        
    except:
        wrong_code.append(Code[i])
        pass
```

```python
df.to_csv('엑셀/소재지별진료.csv',encoding='CP949')
```
